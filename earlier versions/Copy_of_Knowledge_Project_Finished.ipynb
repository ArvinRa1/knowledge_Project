```python
class SampleSat():
    Label = "SampleSat Algorithm"
    def __init__(self, input, probability = 0.5, temperature = 1):
        self.input = sympify(input, evaluate=False)
        self.literals = input.atoms() 
        self.clauses = sympify(self.input.args) 
        self.falseclauses = set()
        self.trueClauses = set()
        self.dictionary = {}
        self.p = probability
        self.t = temperature
    def randomWalk(self):
        """
        Returns the set of false clauses, set of truth clauses, the clause chosen at random from False cluases
        Also returns formula is satisfied when there is no false clauses and the dict of assigned values to vars
        """
        #variables = self.literals
        #clauses = self.clauses
        #dic = self.dictionary
        p = P
        t = T
        vars = list(preprocess(input)[0])
        clauses1 = clauses(input)
        dict1 = dt
        isat = 0
        sat = []

        for atom in vars:   # Randomly initilizing the vars to T and F
            dict1[atom] = random.choice([True, False])
        #dict1[atom]= True  # for testing
        print("clauses1: ",clauses1)
        clauses1 = list(clauses1)
        clF =  set()     # The set that we put the clauses that are still false inside
        clT = clauses1.copy()
        vltf = str()
        for i0 in range(0,1000):
            if len(clF) != 0:
                if vltf[0] in dict1 & vltf[1]:
                    dict1[vltf[0]] = not dict1[vltf[0]]
            for clause in clauses1:   
                temp = clause
                for key in dict1.keys():
                    clT = set(clT)
                    if key in clause.atoms():
                        clause = clause.subs({key: dict1[key]})
                    # Adding false clauses to the set
                    if not clause:
                        clF.add(temp)
                    if temp in clF:
                        clT.remove(temp)
                if len(clF) == 0:
                    sat.append(dict1)
                    print("The formula is satisfied") 
                    break
                if len(clF) != 0:
                    ## Having the hybrid aproach of Heuristic ChooseVar and Metropolis moves
                    if decision(p):
                        vltf = ChooseVar(clT, clF, dict1)
                    else: 
                        vltf = Metropolis(clT, dict1 , T = t)
                    # The set that we put the clauses that are still false inside
                    clF = set()      
                    clT = clauses1.copy()
        return clF, clT, 0, dict1, sat
```

```python
#"  (~a | b) & (~b | ~c) & (~c) & (~a) "

def ChooseVar(sT, clF, dict1):
  clF = clF
  cho1 = random.choice(list(clF))
  sT = sT
  dict2 = dict1
  sT1 = set()
  dict3 = dict2.copy()
  dict4 = dict2.copy()
  bvl = []
  ##### Break value calculation
  for lit in cho1.atoms():
    #print("dict2: ",dict2)
    dict2[lit] = not dict2[lit]
    vars = list(preprocess(input)[0])
    clauses1 = clauses(sT)
    dict1 = dict2
    sat = []
    clauses1 = set(clauses1)
    clF =  set()     # The set that we put the clauses that are still false inside
    clT = clauses1.copy()
    vltf = str()
    for clause in clauses1:   
      temp = clause
      #print("clause before: ", clause)
      for key in dict1.keys():
        clT = set(clT)
        if key in clause.atoms():
          clause = clause.subs({key: dict1[key]})
          #print("clause after: ", clause)              # Adding false clauses to the set
        if not clause:
          clF.add(temp)
        if temp in clF:
          clT.remove(temp)
    #prw1 = PRW(sT, dict2)
    #sT1 = prw1[1]
    #sat = prw1[4]
    #print("sat: ", sat)
    #print("sT1: ",sT1)
    bv = len(sT1) - len(clT)
    bvl.append(max(0, bv))
    dict3.update({lit: [not dict2[lit], max(0, bv)]})
  #####
  #print("dict3",dict3)
  #print("bvl",bvl)
  # chooseVar 
  vltflst = []      # value to flip list 
  probability= 0.5  ##### q probability  
  for key, value in dict3.items():
    if not type(value) == bool :
      if value[1] == 0 :          # if break value equal 0 chose the variable
        vltflst.extend([key, value[1]])
        #print("vltflst: ", vltflst)
        vltf = key              # value to flip 
    else :
      #print("we shouldn't print this")
      if decision(probability):               # with probability of q chose randomly between the vars in cho1
        #print("we shouldn't print this1")
        vltf = random.sample(cho1.atoms(), 1)[0]
      else:                                   # with probability 1-q chose the min of break values in among the literals
        try:
          vltf = vltflst[vltflst.index(min(vltflst))-1]      
          #print("we shouldn't print this2")
        except ValueError:
          pass
        except TypeError:
          pass
  return vltf
```

```python





def Metropolis(sT, dict1, T = 1):
  t = T
  sT = sT
  vars = list(preprocess(input)[0])
  dict1 = dict1
  costl = []
  #if len(dict1) == 0:
  for atom in vars:   # Randomly initilizing the vars to T and F
    dict1[atom] = random.choice([True, False])
      #dict1[atom]= True  # for testing
  dict2 = dict1.copy()
  dict3 = dict2.copy()
  #==============================================================================================================
  ##### Cost calculation
  for lit in vars:
    #print("dict2: ",dict2)
    dict2[lit] = not dict2[lit]
    vars = list(preprocess(input)[0])
    clauses1 = clauses(sT)
    dict1 = dict2
    sat = []
    clauses1 = set(clauses1)
    clF =  set()     # The set that we put the clauses that are still false inside
    clT = clauses1.copy()
    vltf = str()
    clauses1 = clauses(input)
    clauses1 = set(clauses1)
    dict1 = dict2.copy()
    clf =  set()     
    clt = clauses1.copy()
    for clause in clauses1:   
      temp = clause
      #print("clause before: ", clause)
      for key in dict1.keys():
        clt = set(clt)
        if key in clause.atoms():
          clause = clause.subs({key: dict1[key]})
            # Adding false clauses to the set
        if not clause:
          clf.add(temp)
      if temp in clf:
        clt.remove(temp)
    sT1 = clt
    #print("sT1: ", sT1)
    bv = len(sT) - len(sT1)
    costl.append(bv)
    #print("costl: ", costl)
    dict3.update({lit: [not dict2[lit], bv]}) 
  #print("dict3: ",dict3)
  ######==========================================================
  vltf = random.choice(vars)
  costVltf = dict3[vltf][1]
  #print("costVltf: ", costVltf)
  ex = expon.pdf(costVltf) / t
  #print("expon.pdf(costVltf) / t: ", ex)
  rand = random.random()
  #print("rand: ", rand)
  if costVltf <= 0:
    return [vltf, True]
  elif rand  < ex:
    return [vltf , True]
  else:
    return [vltf , not vltf]



def PRW(input, dt = {}, P = 0.5, T = 1):  # Returns the set of false clauses, set of truth clauses, the clause chosen at random from False cluases
                  # Also returns formula is satisfied when there is no false clauses and the dict of assigned values to vars
  p = P
  t = T
  vars = list(preprocess(input)[0])
  clauses1 = clauses(input)
  dict1 = dt
  isat = 0
  sat = []
  if True:#len(dict1) == 0:
    for atom in vars:   # Randomly initilizing the vars to T and F
      dict1[atom] = random.choice([True, False])
      #dict1[atom]= True  # for testing
  #print("clauses1: ",clauses1)
  clauses1 = list(clauses1)
  clF =  set()     # The set that we put the clauses that are still false inside
  clT = clauses1.copy()
  vltf = str()
  i0 =0
  while i0 in range(0,1000):
    #print("i0 : ", i0)
    #print("dict1: ", dict1)
    i0 += 1
    if len(clF) != 0:
      if vltf[0] in dict1 & vltf[1]:
        dict1[vltf[0]] = not dict1[vltf[0]]
    for clause in clauses1:   
      temp = clause
      #print("clause before: ", clause)
      for key in dict1.keys():
        clT = set(clT)
        if key in clause.atoms():
          clause = clause.subs({key: dict1[key]})
          #print("clause after: ", clause)              # Adding false clauses to the set
        if not clause:
          clF.add(temp)
      if temp in clF:
        clT.remove(temp)
    #print("clF: ", clF)
    #print("clT: ", clT)
    if len(clF) == 0:
      sat.append(dict1)
      #print("dict1: ", dict1)
      print("The formula is satisfied") 
      break
    if len(clF) != 0:
      if decision(p):
        vltf = ChooseVar(clT, clF, dict1)
      else: 
        vltf = Metropolis(clT, dict1 , T = t)
      clF = set()      # The set that we put the clauses that are still false inside
      clT = clauses1.copy()
  return clF, clT, 0, dict1, sat
     


#prw = PRW(input)
#print("PRW: ",prw)
#print("prw[1] :",prw[1])
#print(PRW(prw[1],prw[3]))
```
